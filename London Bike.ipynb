## London Bike rides - Data Cleaning
Dataset Reference: https://www.kaggle.com/datasets/hmavrodiev/london-bike-sharing-dataset

**Columns:**
- `timestamp`: grouping time  
- `cnt`: new bike shares count  
- `t1`: temp (°C)  
- `t2`: feels-like temp (°C)  
- `hum`: humidity (%)  
- `wind_speed`: wind (km/h)  
- `weather_code`: weather category  
- `is_holiday`: 1=holiday, 0=not  
- `is_weekend`: 1=weekend, 0=weekday  
- `season`: 0=spring, 1=summer, 2=fall, 3=winter  

`weather_code` category description:
- 1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity
- 2 = scattered clouds / few clouds 
- 3 = Broken clouds
- 4 = Cloudy 
- 7 = Rain/ light Rain shower/ Light rain 
- 10 = rain with thunderstorm 
- 26 = snowfall 
- 94 = Freezing Fog

```python
# Specify new column names
new_cols_name = {
    'timestamp': 'time',
    'cnt': 'count',
    't1': 'temp_actual_C',
    't2': 'temp_feels_C',
    'hum': 'humidity_percent',
    'wind_speed': 'wind_speed_kph',
    'weather_code': 'weather',
    'is_holiday': 'is_holiday',
    'is_weekend': 'is_weekend',
    'season': 'season'
}

# Rename with new names
df_bike.rename(columns=new_cols_name, inplace=True)
df_bike
```

```python
# Change percentage form to decimal form (between 0 to 1)
df_bike.humidity_percent = df_bike.humidity_percent / 100
df_bike
```

```python
# season dictionary
season_name = {
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'fall',
    '3.0': 'winter'
}

# change data type from integer to string
df_bike.season = df_bike.season.astype('str')
# Map 0-3 to actual seasons names
df_bike.season = df_bike.season.map(season_name)

# weather dictionary
weather_name = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

# changing data type from integer to string
df_bike.weather = df_bike.weather.astype('str')
# Map numerical values to the actual weather names
df_bike.weather = df_bike.weather.map(weather_name)
```
